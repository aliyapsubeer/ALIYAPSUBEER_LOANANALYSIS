```python
from sklearn.preprocessing import LabelEncoder

# Convert data from a list to a DataFrame
data_train = pd.DataFrame(data_train)

# Convert the specified columns to strings
cols = ['Gender','Married','Education','Self_Employed','Property_Area']
data_train[cols] = data_train[cols].astype(str)

# Apply label encoding to the specified columns
le = LabelEncoder()
for col in cols:
  data_train[col] = le.fit_transform(data_train[col])

# Display the resulting DataFrame
data_train.head()
```

```python
le = LabelEncoder()
data_train['Loan_Status'] = le.fit_transform(data_train['Loan_Status'])
```

```python
data_train= data_train.fillna(0)
data_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert data from a list to a DataFrame
data_test = pd.DataFrame(data_test)

# Convert the specified columns to strings
cols = ['Gender','Married','Education','Self_Employed','Property_Area']
data_test[cols] = data_test[cols].astype(str)

# Apply label encoding to the specified columns
le = LabelEncoder()
for col in cols:
  data_test[col] = le.fit_transform(data_test[col])

# Display the resulting DataFrame
data_test.head()

```

```python
data_train.dtypes
```

```python

from sklearn.feature_selection import SelectKBest, chi2
X = data_train.drop(['Loan_ID', 'Dependents','Loan_Status'], axis=1)
y = data_train['Loan_Status']
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = model.score(X_test, y_test)
print('Accuracy:', accuracy)
```

```python


# Apply SelectKBest to select the top 2 features
selector = SelectKBest(chi2, k='all')
X_new = selector.fit_transform(X, y)

# Display the selected features
selected_features = X.columns[selector.get_support()]
print('Selected features:', selected_features)


```

```python
from sklearn.tree import DecisionTreeClassifier

# Create a decision tree classifier
model = DecisionTreeClassifier()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = model.score(X_test, y_test)
print('Accuracy:', accuracy)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a random forest classifier
model = RandomForestClassifier()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = model.score(X_test, y_test)
print('Accuracy:', accuracy)
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split


# Create a gradient boosting classifier
model = GradientBoostingClassifier()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = model.score(X_test, y_test)
print('Accuracy:', accuracy)
```
